```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("7.2 data.csv", skiprows = range(10))
```

```python
data.head(5)
```

```python
x = np.array(data.iloc[:,0])
y1 = np.array(data.iloc[:,1])
y2 = np.array(data.iloc[:,2])
```

# 1. Line graph

```python
#Using matplotlib
plt.plot(x, y2, label = 'Antartica')
plt.plot(x, y1, label = 'Greenland')
plt.ylabel('Land ice mass anomaly (Gt)')
plt.xlabel('Time (year)')
plt.title('Land ice mass from 2002 to 2015')
plt.legend()
plt.grid(axis = 'y')
plt.show()
```

```python
#using seaborn
sns.set_style("whitegrid")
fig = sns.lineplot(x,y1, label = 'Greenland')
fig = sns.lineplot(x,y2, label = 'Antarctica')
fig.set(xlabel = 'Time(year)', ylabel = 'Land ice mass anomaly (Gt)', title = 'Greenland and Antarctica land ice mass through time')
plt.legend()
plt.show()
```

Student comment: From these graph we can see that there is a general decreasing trend in the land ice mass in both Greenland and Antarctica. Even though the data fluctuate every year, the trend is apparent and is a concrete proof of temperature increase in both regions.

# 2. Scatterplot

```python
#Using matplotlib
plt.scatter(y1, y2, c = x)
plt.ylabel('Greenland land ice mass anomaly (Gt)')
plt.xlabel('Antarctica land ice mass anomaly (Gt)')
plt.title('Land ice mass from 2002 to 2015')
plt.show()
```

```python
#using seaborn
sns.set()
fig = sns.scatterplot(y1, y2, hue = x, color = 'green')
fig.set(xlabel = 'Greenland ice mass anomaly (Gt)', ylabel = 'Antartica ice mass anomaly (Gt)', title = 'Greenland and Antartica ice mass through time')
plt.show()
```

These graphs show the correlation between Greenland and Antarctica land ice mass. It can be determined that they are strongly correlated.

# 3. Histogram

```python
plt.hist(data.iloc[:,1], bins = 10, label = 'Greenland')
plt.hist(data.iloc[:,2], bins = 10, color = 'r', label = 'Antarctica')
plt.ylabel('Frequency')
plt.xlabel('Land ice mass')
plt.title('Histogram of land ice mass')
plt.legend()
plt.show()
```

```python
fig = sns.distplot(y1, bins = 10, label = 'Greenland')
fig = sns.distplot(y2, bins = 10, label = 'Antarctica')
fig.set(xlabel = 'Greenland ice mass anomaly (Gt)', ylabel = 'frequency density', title = 'Greenland and Antarctica land ice mass distribution')
plt.legend()
plt.show()
```

These histogram shows the distribution of the data. From this we can see that the Antarctica has a smaller standard deviation in compared to Greenland. However it does not provide us with much insight into our topic on global warming.
